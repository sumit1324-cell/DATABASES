**QUESTION  3**. Display All Employee Records Sorted by Salary (Lowest to Highest) Hint: Use the ORDER BY clause on the salary column.

**ANSWER**

```python
SELECT *
FROM employees
ORDER BY salary ASC;
```

This query displays all employee records from the employees table sorted by salary in ascending order (lowest to highest).

**QUESTION  4**. Show Employees Sorted by Department (AZ) and Salary (High  Low)

**ANSWER**


```python
SELECT *
FROM employees
ORDER BY department ASC, salary DESC;
```

This query displays employee records sorted first by department in alphabetical order (AZ) and then by salary from highest to lowest within each department.

Q5. List All Employees in the IT Department, Ordered by Hire Date (Newest First)



**ANSWER**

```python
SELECT *
FROM employees
WHERE department = 'IT'
ORDER BY hire_date DESC;
```

This query displays all employees working in the IT department, sorted by hire date from newest to oldest.

**QUESTION  9.** What is the Difference Between a Primary Key and a Foreign Key?

Difference Between Primary Key and Foreign Key
Primary Key

A primary key uniquely identifies each record in a table

It cannot contain NULL values

Each table can have only one primary key

It ensures uniqueness and data integrity

Example:
employee_id in the employees table

Foreign Key

A foreign key is used to link one table with another table

It can contain duplicate values

It can contain NULL values

A table can have more than one foreign key

Example:
employee_id in another table referencing the employees table

**QUESTION 10.**
What Are Constraints in SQL and Why Are They Used?

**ANSWER**

Constraints in SQL
Meaning of Constraints

Constraints in SQL are rules applied to table columns to control the type of data that can be stored in a table.
They help maintain the accuracy and integrity of data in the database.



---



Why Constraints Are Used

Constraints are used to:

->Prevent invalid data from being inserted

->Ensure data consistency and reliability

->Maintain data integrity

->Enforce business rules at the database level

---

Common Types of Constraints

->PRIMARY KEY  Uniquely identifies each record

->FOREIGN KEY  Links tables and maintains relationships

->NOT NULL  Ensures a column cannot have NULL values

->UNIQUE  Prevents duplicate values

->CHECK  Ensures values meet a specific condition

->DEFAULT  Assigns a default value if none is provided

---

Answer

Constraints in SQL are rules that restrict the type of data stored in a table. They are used to maintain data accuracy, consistency, and integrity in a database.
